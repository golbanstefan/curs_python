# 🔁 Bucluri (Loops) în Python

Bucla este un concept esențial în programare care îți permite să repeți un bloc de cod de mai multe ori. Python oferă două tipuri principale: `for` și `while`.

## 🔄 Bucla `for`
- Iterare prin liste, tuple sau șiruri
- Utilizarea funcției `range()` pentru generarea unei secvențe
- `enumerate()` oferă indexul și valoarea în timpul iterării

## 🔁 Bucla `while`
- Continuă cât timp condiția este `True`
- Asigură-te că bucla se va opri la un moment dat

## 🧭 Controlul buclei
- `break`: întrerupe complet bucla
- `continue`: trece la următoarea iterație
- `pass`: nu face nimic (folosit ca placeholder)

```python
# Exemplu: bucle și instrucțiuni de control

print("Iterare printr-o listă:")
for indicator in ["Ocupare", "Educație", "Venit"]:
    print(indicator)

print("Utilizare range():")
for i in range(1, 6):
    print(f"Număr: {i}")

print("Enumerare listă:")
for index, value in enumerate(["Alfa", "Beta", "Gamma"]):
    print(f"Index {index}: {value}")

print("Buclă while:")
count = 5
while count > 0:
    print(f"Numărătoare inversă: {count}")
    count -= 1

print("Controlul buclei:")
for num in range(1, 10):
    if num == 5:
        break
    elif num % 2 == 0:
        continue
    else:
        pass
    print(f"Număr impar: {num}")
```
