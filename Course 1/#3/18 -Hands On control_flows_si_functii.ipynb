# 🔁 Controlul Fluxului și Funcțiile în Python

În Python, controlul fluxului definește ordinea în care este executat codul. Acesta permite rularea condiționată a blocurilor de cod și repetarea anumitor acțiuni. Funcțiile, pe de altă parte, permit gruparea codului în unități reutilizabile pentru o organizare mai bună.

## ✅ Instrucțiuni Condiționale
- `if`: execută un bloc de cod dacă o condiție este adevărată
- `if-else`: adaugă o alternativă dacă condiția este falsă
- `if-elif-else`: verifică mai multe condiții în ordine

```python
# Exemplu: Condiționale
temperatura = 25

if temperatura > 30:
    print("Este foarte cald afară!")
elif temperatura > 20:
    print("Este cald afară.")
else:
    print("Este răcoare afară.")
```

## 🔁 Bucles (Loop-uri)
- `for`: parcurge o listă sau șir
- `while`: execută un bloc cât timp o condiție este adevărată

```python
# For loop
fructe = ["mar", "banana", "cireșe"]
for fruct in fructe:
    print(fruct)

# While loop
count = 0
while count < 3:
    print("Count:", count)
    count += 1
```

## 🔧 Alte Instrucțiuni de Control
- `break`: întrerupe bucla
- `continue`: trece la următoarea iterație
- `pass`: nu face nimic, doar completează sintaxa

## 🧭 Funcții în Python
- Se definesc cu `def`
- Se pot apela oricând după definire

```python
# Exemplu: funcție simplă
def salut(nume):
    print("Salut, " + nume + "!")

salut("Alice")
```

```python
# Exemplu: funcție cu return
def adunare(a, b):
    return a + b

rezultat = adunare(3, 4)
print("Suma este:", rezultat)
```
