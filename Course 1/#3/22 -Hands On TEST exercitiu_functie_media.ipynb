# 🧠 Exercițiu Python: Funcția pentru calculul mediei

Creează o funcție care calculează media unei liste de numere.

## Pași:
1. Definește o funcție numită `calculeaza_media` care primește o listă de numere.
2. Verifică dacă lista este goală. Dacă da, returnează `0`.
3. Calculează suma numerelor.
4. Împarte suma la lungimea listei.
5. Returnează media calculată.
6. Apelează funcția cu o listă de test și afișează rezultatul.

```python
# Pasul 1: Definește funcția
def calculeaza_media(lista):
    # Pasul 2: Verificare listă goală
    if not lista:
        return 0
    # Pasul 3: Suma elementelor
    suma = sum(lista)
    # Pasul 4: Calcul media
    media = suma / len(lista)
    # Pasul 5: Returnează rezultatul
    return media

# Pasul 6: Apel funcție cu listă de test
rezultat = calculeaza_media([10, 20, 30, 40])
print(f"Media este: {rezultat}")
```
