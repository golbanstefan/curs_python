# Transfromati o lista bidimensionalÄƒ itrun vector

```python
# ğŸ§­ Exemplu de listÄƒ bidimensionalÄƒ
matrice = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

my_vector=[]

for row in matrice:
    for item in row:
       my_vector.append(item)

print(my_vector)
```
