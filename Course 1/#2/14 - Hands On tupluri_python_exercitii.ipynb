Tuplurile (`tuple`) sunt secvenÈ›e imuabile folosite pentru a stoca colecÈ›ii de date. Sunt utile atunci cÃ¢nd datele nu trebuie modificate.

## ðŸ”‘ Caracteristici ale tuplurilor
- **Imuabile:** Nu pot fi modificate dupÄƒ creare.
- **Ordonate:** PÄƒstreazÄƒ ordinea elementelor È™i permit indexare/slicing.
- **Flexibile:** Pot conÈ›ine tipuri de date diferite, inclusiv alte tupluri.
- **Eficiente:** Mai rapide È™i mai puÈ›in consumatoare de memorie faÈ›Äƒ de liste.
- **Destructurare:** Se pot atribui variabile direct (`x, y = coordonate`).
- **OperaÈ›ii suportate:** concatenare, test de apartenenÈ›Äƒ, repetare etc.

## ðŸ“Œ CÃ¢nd sÄƒ foloseÈ™ti tupluri?
- Pentru date constante precum coordonate, coduri RGB, etc.
- Ca chei Ã®n dicÈ›ionare (fiind imuabile).
- Pentru a returna mai multe valori dintr-o funcÈ›ie.

```python
# ðŸ§ª Exemple cu tupluri
coordonate = (10.0, 20.0)
culori = ("roÈ™u", "verde", "albastru")

# Destructurare
x, y = coordonate

# Accesare
prima_culoare = culori[0]

# Concatenare
combinat = coordonate + (30.0,)

# ApartenenÈ›Äƒ
exista_albastru = "albastru" in culori

# Tupluri imbricate
tuplu_imbricat = ((1, 2), (3, 4))

print("Coordonate:", coordonate)
print("X:", x, "Y:", y)
print("Prima culoare:", prima_culoare)
print("Combinat:", combinat)
print("'albastru' existÄƒ Ã®n culori?:", exista_albastru)
print("Tuplu imbricat:", tuplu_imbricat)
```
