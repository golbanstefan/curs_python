# Comentarii în Python

Comentariile în Python sunt utilizate pentru a explica codul și pentru a-l face mai ușor de înțeles. Ele sunt ignorate de interpretul Python în timpul execuției.

## 1. Comentarii pe o singură linie

- Comentariile pe o singură linie încep cu caracterul `#`.
- Ele pot fi scrise pe o linie separată sau la finalul unei instrucțiuni de cod.

```python
# Acesta este un comentariu pe o singură linie
print("Salut lume!")  # Acest comentariu urmează o linie de cod
```

## 2. Comentarii pe mai multe linii

- Comentariile pe mai multe linii pot fi scrise între triple ghilimele: `'''` sau `"""`.
- Tehnic vorbind, acestea sunt șiruri de caractere multilini, nu comentarii propriu-zise, dar sunt folosite adesea în acest scop.

```python
'''
Aceasta este o formă de comentariu
pe mai multe linii folosind ghilimele simple triple.
Aceasta poate fi folosită pentru documentare sau pentru a dezactiva cod temporar.
'''

"""
Aceasta este o altă metodă de a scrie comentarii multilini
folosind ghilimele duble triple.
"""
```

## 3. Practici Recomandate

- Folosește comentarii pentru a explica logica complexă sau neclară a codului.
- Comentariile trebuie să fie concise și clare.
- Evită comentariile evidente (`x = 5  # Setează x la 5`).
- Menține comentariile actualizate — comentariile învechite pot induce în eroare.
- Folosește docstring-uri (comentarii între triple ghilimele) pentru documentarea funcțiilor, claselor și modulelor.

## 4. Notă Finală

Comentariile coerente și semnificative îi ajută pe ceilalți (și pe tine în viitor) să înțeleagă codul mai ușor. Ele sunt o componentă esențială a scrierii unui software ușor de întreținut.
