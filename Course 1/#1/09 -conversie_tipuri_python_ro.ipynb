# Conversia Tipurilor de Date în Python

Conversia tipurilor de date este procesul de transformare a unei valori dintr-un tip în altul. Există două tipuri principale:

- **Conversie implicită** (automată): Python convertește automat un tip într-altul dacă este sigur (ex: `int` → `float`).
- **Conversie explicită** (manuală): Se realizează folosind funcții precum `int()`, `str()`, `float()`, etc.

---

## Conversie Implicită

```python
a = 10        # int
b = 2.5       # float
c = a + b     # c devine float
print(c, type(c))
# Output: 12.5 <class 'float'>
```

## Conversie Explicită (Cast)

```python
# String → int și float
num_str = "123"
num_int = int(num_str)
num_float = float(num_str)
print(num_int, type(num_int))
print(num_float, type(num_float))
```

```python
# Int/Float → String
x = 42
x_str = str(x)
print("Valoarea este " + x_str)
```

```python
# Listă de numere → listă de stringuri
nums = [1, 2, 3]
str_nums = list(map(str, nums))
print(str_nums)
```

## Conversii Frecvente

| Din \ Către | `int()` | `float()` | `str()` | `bool()` |
|-------------|---------|-----------|----------|----------|
| `str`       | Da (dacă conține număr valid) | Da | Da | `bool('abc') → True` |
| `int`       | -       | Da        | Da       | `bool(0) → False` |
| `float`     | Da (pierde zecimalele) | - | Da | `bool(0.0) → False` |
| `bool`      | Da (`True → 1`) | Da (`False → 0.0`) | Da | - |

### Notă:
- `int("12.5")` va genera eroare, deoarece `"12.5"` nu este un întreg valid.
- `bool()` returnează `False` doar pentru valori considerate „goale”: `0`, `0.0`, `""`, `[]`, `None`, etc.

---

## Exemple Practice


```python
# Conversie de la string la boolean
print(bool(""))     # False
print(bool("abc"))  # True

# Conversie de la float la int
print(int(12.9))     # 12 (trunchiere)

# Verificare tip după conversie
x = str(100)
print(x, type(x))
```
