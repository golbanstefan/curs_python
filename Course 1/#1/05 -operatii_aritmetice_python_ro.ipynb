# Operații Aritmetice în Python

Operațiile aritmetice sunt fundamentale în Python și ne permit să efectuăm calcule numerice. Mai jos sunt cele mai utilizate operații.

## Operații Aritmetice de Bază

- **Adunare (`+`)**: Adună două numere.
- **Scădere (`-`)**: Scade al doilea număr din primul.
- **Înmulțire (`*`)**: Înmulțește două numere.
- **Împărțire (`/`)**: Împarte primul număr la al doilea și returnează un float.
- **Împărțire întreagă (`//`)**: Returnează câtul întreg al împărțirii.
- **Modul (`%`)**: Returnează restul împărțirii.
- **Exponențiere (`**`)**: Ridică primul număr la puterea celui de-al doilea.

### Exemplu de bază

```python
# Definim două numere
a = 15
b = 4

# Operații de bază
print("Adunare:", a + b)
print("Scădere:", a - b)
print("Înmulțire:", a * b)
print("Împărțire:", a / b)
print("Împărțire întreagă:", a // b)
print("Modul:", a % b)
print("Putere:", a ** b)
```

## Exemple Suplimentare

### Operații combinate și prioritatea operatorilor

```python
# Se respectă ordinea operațiilor (PEMDAS)
rezultat = 10 + 2 * 3
print("10 + 2 * 3 =", rezultat)

# Se pot folosi paranteze pentru a schimba ordinea
rezultat = (10 + 2) * 3
print("(10 + 2) * 3 =", rezultat)
```

### Operații cu float și int

```python
# Împărțire între float și int
x = 7.5
y = 2
print("Împărțire:", x / y)
print("Împărțire întreagă:", x // y)
```

### Exponențiere și radical

```python
# Exponențiere și extragerea rădăcinii pătrate
print("2 la puterea 5:", 2 ** 5)
print("Rădăcina pătrată din 16:", 16 ** 0.5)
```

### Negarea numerelor

```python
# Lucrul cu valori negative
a = -10
b = 4
print("a % b =", a % b)
print("a // b =", a // b)
```

### Incrementare și decrementare

```python
# Python nu are ++ sau --, dar putem adăuga sau scădea direct
numar = 10
numar += 1  # echivalent cu numar = numar + 1
print("Incrementat:", numar)

numar -= 2
print("Decrementat:", numar)
```

## Observații

- Împărțirea cu `/` returnează întotdeauna un `float`, chiar dacă este exactă.
- Operatorii pot fi combinați în expresii complexe.
- Împărțirea întreagă și modulul pot da rezultate diferite cu numere negative.
