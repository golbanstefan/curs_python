# Sintaxă de bază și variabile în Python

Variabilele în Python acționează ca recipiente pentru stocarea valorilor de date. Python este un limbaj tipat dinamic, ceea ce înseamnă că tipul unei variabile este dedus automat pe baza valorii atribuite.

## Puncte cheie

Variabilele în Python sunt folosite pentru a stoca valori de date. Spre deosebire de alte limbaje de programare, Python nu necesită declararea explicită a tipului variabilei; tipul este dedus automat în funcție de valoarea atribuită. Acest lucru face atribuirea variabilelor simplă și flexibilă.

## Caracteristici principale ale variabilelor în Python

- **Tipare dinamică:** Tipul este atribuit automat în funcție de valoare.
- **Tipuri de date flexibile:** Variabilele pot stoca întregi, șiruri de caractere, zecimale etc.
- **Reatribuire:** Variabilele pot fi reatribuite cu alte valori sau chiar tipuri diferite.

- Variabilele pot stoca date de diferite tipuri: întregi, șiruri de caractere, zecimale.
- Python permite reatribuirea variabilelor cu un alt tip de date.
- Operatorul `=` este folosit pentru a atribui valori variabilelor.

## Exemplu pas cu pas

### 1. Atribuirea unei valori întregi

```python
# Atribuirea unei valori întregi
varsta = 28
print(varsta)
```

### 2. Atribuirea unei valori de tip șir de caractere

```python
# Atribuirea unei valori de tip string
nume = "Maria"
print(nume)
```

### 3. Atribuirea unei valori zecimale

```python
# Atribuirea unei valori de tip float
inaltime = 1.65
print(inaltime)
```

### 4. Afișarea tuturor variabilelor

```python
# Afișarea tuturor valorilor
print(varsta, nume, inaltime)
```

### 5. Reatribuirea variabilelor cu alt tip

```python
# Reatribuirea variabilei 'varsta' cu un string
varsta = "douăzeci și opt"
print(varsta)
```

## Bune practici

- Alege nume semnificative pentru variabile pentru o lizibilitate mai bună, ex: `varsta` în loc de `v`.
- Numele variabilelor sunt sensibile la majuscule: `Varsta` și `varsta` sunt diferite.
- Evită utilizarea cuvintelor rezervate de Python ca nume de variabile (ex: `if`, `True`, `None`).
- Urmează convențiile de denumire în Python:
  - Folosește `snake_case` pentru numele variabilelor, ex: `nume_utilizator`.
  - Începe cu o literă sau underscore, ex: `_privat`.

## Observații suplimentare

Tiparea dinamică a Python îți permite să creezi variabile fără a declara explicit tipul lor, dar această flexibilitate implică o responsabilitate crescută pentru alegerea unor nume clare și organizarea atentă a codului pentru a menține claritatea și evitarea erorilor.
