## Tipuri de celule și moduri de celule
- Tipuri de celule: **Code** și **Markdown**. Se selectează din lista din bara de instrumente.
- Moduri de celule: **Edit Mode** (margine verde) sau **Command Mode** (margine albastră).
- Dublu click într-o celulă Markdown sau click într-o celulă de cod pentru a intra în **Edit Mode**.
- Apasă tasta Escape pentru a intra în **Command Mode**.
- **Command Mode** este folosit pentru navigare, ștergere sau mutare celule.
- În modul comandă, apasă "m" pentru Markdown și "y" pentru celulă de cod.


## Executarea celulelor de cod

- Trebuie să „executăm” o celulă pentru a rula codul Python din ea.
- Apasă **Shift + Enter** pentru a executa celula și a trece la următoarea.
- Apasă **Ctrl + Enter** pentru a executa celula și a rămâne în aceeași celulă.
- Executarea unei celule Markdown va afișa formatarea.
- O celulă poate fi executată de oricâte ori.
- Celulele pot fi executate în orice ordine.


- Mai multe operații pot fi efectuate într-o singură celulă.
- Celula va afișa automat rezultatul ultimei expresii.
- Orice operație care nu este salvată într-o variabilă va fi pierdută.


## Importul bibliotecilor
- O bibliotecă sau un modul Python este un set de funcționalități ce adaugă capabilități noi.
- Bibliotecile nu sunt importate automat pentru a economisi memorie — se folosesc la nevoie.
- Bibliotecile trebuie importate în fiecare proiect (Notebook).
- Importă o bibliotecă folosind cuvântul cheie **`import`** urmat de numele bibliotecii.
- Bibliotecile pot avea un alias, un nume alternativ. Aliasurile sunt opționale.
- Un alias reduce caracterele scrise când facem referire la bibliotecă.
- Aliasul poate avea orice nume; standardul comunității pentru `pandas` este `pd`.
- După `import` folosește `as` și aliasul (fără ghilimele).
- Mai multe biblioteci pot fi importate într-o singură celulă.
- Biblioteca `pandas` este construită peste biblioteca `numpy` (Numerical Python).
- Biblioteca `numpy` poate fi folosită pentru a genera seturi de date false — o vom folosi ocazional.
