# Tipuri de Date în Python

Python oferă mai multe tipuri de date predefinite care clasifică tipul de valoare pe care o variabilă o poate stoca. Mai jos sunt cele mai comune tipuri de date.

## Tipuri Numerice

**Tipuri numerice** sunt folosite pentru a reprezenta numere:

- **int**: Reprezintă numere întregi, ex: `10`, `-5`.
- **float**: Reprezintă numere zecimale, ex: `3.14`, `-0.5`.

```python
# Tipuri numerice
numar_intreg = 10       # int
numar_zecimal = 3.14    # float

print(type(numar_intreg))   # <class 'int'>
print(type(numar_zecimal))  # <class 'float'>
```

## Tip Text (șir de caractere - str)

Textul în Python este reprezentat prin tipul **str** (string):

- Șirurile sunt delimitate prin ghilimele simple `'` sau duble `"`.

```python
# Tip text
salut = "Salut, lume!"
print(type(salut))  # <class 'str'>
```

## Tip Boolean

Tipul **bool** reprezintă valori logice: `True` sau `False`.

- Sunt utilizate în decizii și condiții.

```python
# Tip boolean
activ = True
inchis = False

print(type(activ))  # <class 'bool'>
print(inchis)       # False
```

## Conversii și combinații între tipuri

```python
# Adăugare între int și float
rezultat = 10 + 3.5
print(rezultat)         # 13.5
print(type(rezultat))   # <class 'float'>
```

```python
# Încercare de adăugare între int și string (va genera eroare)
numar = 5
text = "5"
print(numar + text)  # TypeError: unsupported operand type(s)
```

```python
# Încercare de adăugare între int și string (va genera eroare)
numar = 5
text = "5"
print(text + numar)  # TypeError: unsupported operand type(s)
```

```python
# Conversie manuală a string-ului în int pentru a adăuga
rezultat = numar + int(text)
print(rezultat)  # 10
```

```python
# Convertim un număr într-un string și concatenăm
rezultat_text = str(numar) + text
print(rezultat_text)  # '55'
```

```python
# Comparăm valori booleene cu alte tipuri
print(True + 1)  # 2, deoarece True este echivalent cu 1
print(False + 10)  # 10, deoarece False este echivalent cu 0
```

> **Notă:** Python permite conversii între anumite tipuri de date, dar în unele cazuri trebuie făcute explicit (manual), altfel se generează erori.
