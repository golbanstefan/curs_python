# Formatarea șirurilor în Python

Python oferă mai multe modalități de a insera sau afișa valori într-un șir de caractere. Mai jos sunt prezentate cele mai utilizate metode.

## 1. Folosind f-string
F-string-urile (format string literals) au fost introduse în Python 3.6 și permit interpolarea variabilelor direct în șiruri.

```python
name = "Ana"
age = 25

print(f"Numele meu este {name} și am {age} ani.")
```

## 2. Folosind metoda `.format()`
Metoda `.format()` înlocuiește acoladele din șir cu valorile transmise.

```python
name = "Ion"
age = 40

print("Mă numesc {} și am {} ani.".format(name, age))
```

## 3. Folosind formatul cu `%`
Aceasta este o metodă mai veche dar încă suportată. Se folosește `%s`, `%d`, etc.

```python
name = "Maria"
age = 35

print("Numele meu este %s și am %d ani." % (name, age))
```

## 4. Formatarea numerelor
Putem controla câte zecimale afișăm și alinierea valorilor.

```python
pi = 3.1415926535
print(f"Pi cu 2 zecimale: {pi:.2f}")
print("Pi cu 4 zecimale: {:.4f}".format(pi))
```

## 5. Aliniere și umplere
Putem alinia textul la stânga, dreapta sau centru și umple cu un caracter specific.

```python
text = "Salut"
print(f"|{text:<10}|")   # aliniat la stânga
print(f"|{text:>10}|")   # aliniat la dreapta
print(f"|{text:^10}|")   # centrat
print(f"|{text:*^10}|")  # centrat și umplut cu *
```

## 6. Interpolarea expresiilor
Putem evalua expresii direct în f-string.

```python
a = 5
b = 3
print(f"Rezultatul lui {a} + {b} este {a + b}")
```

## Observație
F-string este metoda recomandată începând cu Python 3.6 deoarece este mai eficientă, mai lizibilă și mai concisă.
