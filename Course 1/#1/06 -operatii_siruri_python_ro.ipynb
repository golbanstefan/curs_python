# Operații cu șiruri de caractere în Python

Șirurile de caractere (strings) sunt secvențe imutabile de caractere în Python. Acestea sunt foarte utile pentru procesarea textului și oferă o gamă variată de operații.

## Concatenare
Concatenarea se face cu operatorul `+`, care unește două sau mai multe șiruri de caractere.

```python
prenume = "Ion"
nume = "Popescu"
nume_complet = prenume + " " + nume
print("Nume complet:", nume_complet)
```

## Repetare
Putem repeta un șir folosind operatorul `*`.

```python
ras = "ha"
ras_repetat = ras * 4
print("Rezultat:", ras_repetat)
```

## Indexare și slicing
Putem accesa caractere individuale sau subșiruri folosind indici.

```python
text = "Programare"
primul = text[0]         # 'P'
ultimul = text[-1]       # 'e'
subsir = text[3:]       # 'grama'
print("Primul caracter:", primul)
print("Ultimul caracter:", ultimul)
print("Subșir (3:8):", subsir)
```

## Funcții utile pentru stringuri

```python
mesaj = "salut lume"
print("Upper:", mesaj.upper())
print("Capitalize:", mesaj.capitalize())
print("Este alfabetic:", mesaj.isalpha())
print("Split:", mesaj.split())
```

## Formatarea șirurilor
Putem înlocui valori într-un șir folosind metoda `format()` sau f-string.

```python
nume = "Elena"
varsta = 25
print("Numele este {} și are {} ani.".format(nume, varsta))
print(f"Numele este {nume} și are {varsta} ani.")
```

## Conversii între tipuri
Putem converti alte tipuri la string cu `str()`, dar trebuie să convertim explicit când dorim concatenare.

```python
x = 10
text = "Numărul este " + str(x)
print(text)

# conversie eșuată fără str()
# print("Numărul este " + x)  # va da eroare
```

## Verificarea apartenenței
Putem verifica dacă un șir conține un alt șir cu operatorul `in`.

```python
propozitie = "Python este distractiv"
print("python" in propozitie)  # True
print("java" in propozitie)    # False
```
