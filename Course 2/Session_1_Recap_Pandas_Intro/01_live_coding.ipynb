### 1.1 Date structurate cu liste È™i dicÈ›ionare

````python
# Exemplu: datele din homework 1 - dicÈ›ionare Ã®n listÄƒ
regiuni_dict = [
    {
        "regiune": "ChiÈ™inÄƒu",
        "populatie": 532513,
        "rata_ocupare": 67.5,
        "venit_mediu": 8750,
        "rata_alfabetizare": 99.8,
        "sector_dominant": "Servicii",
        "numar_intreprinderi": 15420,
        "rata_somaj": 5.2
    },
    {
        "regiune": "BÄƒlÈ›i",
        "populatie": 102457,
        "rata_ocupare": 63.2,
        "venit_mediu": 6420,
        "rata_alfabetizare": 98.5,
        "sector_dominant": "Industrie",
        "numar_intreprinderi": 3840,
        "rata_somaj": 6.8
    },
    {
        "regiune": "Cahul",
        "populatie": 28763,
        "rata_ocupare": 59.8,
        "venit_mediu": 5340,
        "rata_alfabetizare": 97.2,
        "sector_dominant": "AgriculturÄƒ",
        "numar_intreprinderi": 1250,
        "rata_somaj": 8.4
    }
]

print("ğŸ“Š Date regiuni (format dicÈ›ionare):")
print(f"Tip: {type(regiuni_dict)}")
print(f"NumÄƒr regiuni: {len(regiuni_dict)}")
print(f"\nPrima regiune:\n{regiuni_dict[0]}")
````

### 1.2 OperaÈ›ii clasice din Curs 1

````python
# Calcul populaÈ›ie totalÄƒ - metoda din Curs 1
def calculeaza_populatie_totala(lista_regiuni):
    """CalculeazÄƒ populaÈ›ia totalÄƒ din toate regiunile."""
    total = 0
    for regiune in lista_regiuni:
        total += regiune["populatie"]
    return total

populatie_totala = calculeaza_populatie_totala(regiuni_dict)
print(f"PopulaÈ›ie totalÄƒ: {populatie_totala:,} locuitori")
````

````python
# Filtrare - gÄƒseÈ™te regiuni cu populaÈ›ie > 50,000
def filtreaza_regiuni_mari(lista_regiuni, prag=50000):
    """ReturneazÄƒ regiunile cu populaÈ›ie peste pragul specificat."""
    regiuni_mari = []
    for regiune in lista_regiuni:
        if regiune["populatie"] > prag:
            regiuni_mari.append(regiune)
    return regiuni_mari

regiuni_mari = filtreaza_regiuni_mari(regiuni_dict)
print(f"\nğŸ™ï¸ Regiuni mari (>50,000 locuitori): {len(regiuni_mari)}")
for reg in regiuni_mari:
    print(f"  - {reg['regiune']}: {reg['populatie']:,}")
````

````python
# Sortare - ordoneazÄƒ dupÄƒ venit mediu
def sorteaza_dupa_venit(lista_regiuni):
    """SorteazÄƒ regiunile descrescÄƒtor dupÄƒ venit mediu."""
    return sorted(lista_regiuni, key=lambda x: x["venit_mediu"], reverse=True)

regiuni_sortate = sorteaza_dupa_venit(regiuni_dict)
print("\nğŸ’° Top regiuni dupÄƒ venit mediu:")
for i, reg in enumerate(regiuni_sortate, 1):
    print(f"  {i}. {reg['regiune']}: {reg['venit_mediu']} lei")
````

### Task: CalculaÈ›i venitul mediu naÈ›ional

````python
print("=" * 60)
print("TASK: CalculaÈ›i venitul mediu naÈ›ional")
print("=" * 60)

# METODA CURS 1: ListÄƒ de dicÈ›ionare
print("\nğŸ“ METODA 1: Curs 1 (liste + dicÈ›ionare)")
print("-" * 60)

def venit_mediu_national_vechi(lista_regiuni):
    """CalculeazÄƒ venitul mediu naÈ›ional (metoda Curs 1)."""
    suma = 0
    count = 0
    for regiune in lista_regiuni:
        suma += regiune['venit_mediu']
        count += 1
    return suma / count

# Trebuie sÄƒ citim È™i parsÄƒm CSV manual
import csv
with open('../datasets/regiuni_moldova.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    regiuni_list = []
    for row in reader:
        regiuni_list.append({
            'regiune': row['regiune'],
            'populatie': int(row['populatie']),
            'rata_ocupare': float(row['rata_ocupare']),
            'venit_mediu': float(row['venit_mediu']),
            'rata_alfabetizare': float(row['rata_alfabetizare']),
            'sector_dominant': row['sector_dominant'],
            'numar_intreprinderi': int(row['numar_intreprinderi']),
            'rata_somaj': float(row['rata_somaj'])
        })

venit_vechi = venit_mediu_national_vechi(regiuni_list)
print(f"Venit mediu: {venit_vechi:.2f} lei")
print("Linii de cod: ~25 (inclusiv citirea CSV)")

# METODA PANDAS
print("\nâš¡ METODA 2: Pandas")
print("-" * 60)

# ReÃ®ncÄƒrcÄƒm DataFrame-ul pentru claritate
df = pd.read_csv('../datasets/regiuni_moldova.csv')
venit_nou = df['venit_mediu'].mean()

print(f"Venit mediu: {venit_nou:.2f} lei")
print("Linii de cod: 2")

print("\n" + "=" * 60)
print("âœ… REZULTAT IDENTIC, dar pandas e ~12x mai concis!")
print("ğŸ’¡ Pandas gestioneazÄƒ automat conversiile de tip È™i citirea CSV")
print("=" * 60)
````

### Task: GÄƒsiÈ›i regiunea cu cel mai mare numÄƒr de Ã®ntreprinderi

````python
print("=" * 60)
print("TASK: Regiunea cu cele mai multe Ã®ntreprinderi")
print("=" * 60)

# AsigurÄƒm cÄƒ regiuni_list existÄƒ È™i are toate cÃ¢mpurile necesare
# (Ã®n cazul Ã®n care celula 34 nu a fost rulatÄƒ)
if ('regiuni_list' not in globals() or 
    not regiuni_list or 
    'numar_intreprinderi' not in regiuni_list[0]):
    import csv
    with open('../datasets/regiuni_moldova.csv', 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        regiuni_list = []
        for row in reader:
            regiuni_list.append({
                'regiune': row['regiune'],
                'populatie': int(row['populatie']),
                'rata_ocupare': float(row['rata_ocupare']),
                'venit_mediu': float(row['venit_mediu']),
                'rata_alfabetizare': float(row['rata_alfabetizare']),
                'sector_dominant': row['sector_dominant'],
                'numar_intreprinderi': int(row['numar_intreprinderi']),
                'rata_somaj': float(row['rata_somaj'])
            })

# CURS 1
print("\nğŸ“ METODA CURS 1:")
max_intreprinderi = 0
regiune_top = ""
for regiune in regiuni_list:
    if regiune['numar_intreprinderi'] > max_intreprinderi:
        max_intreprinderi = regiune['numar_intreprinderi']
        regiune_top = regiune['regiune']

print(f"Regiunea: {regiune_top}")
print(f"Ãntreprinderi: {max_intreprinderi:,}")
print("Linii de cod: ~8")

# PANDAS
print("\nâš¡ METODA PANDAS:")
idx_max = df['numar_intreprinderi'].idxmax()
regiune_max = df.loc[idx_max]

print(f"Regiunea: {regiune_max['regiune']}")
print(f"Ãntreprinderi: {regiune_max['numar_intreprinderi']:,}")
print("Linii de cod: 2")

# SAU mai simplu:
print("\nâš¡ VARIANTÄ‚ ALTERNATIVÄ‚ (1 linie):")
print(df.loc[df['numar_intreprinderi'].idxmax(), ['regiune', 'numar_intreprinderi']])
````

### Task: GrupaÈ›i regiunile dupÄƒ sectorul dominant È™i calculaÈ›i populaÈ›ia totalÄƒ per sector

````python
print("=" * 60)
print("TASK: PopulaÈ›ie totalÄƒ per sector dominant")
print("=" * 60)

# AsigurÄƒm cÄƒ regiuni_list existÄƒ È™i are toate cÃ¢mpurile necesare
# (Ã®n cazul Ã®n care celula 34 nu a fost rulatÄƒ)
if ('regiuni_list' not in globals() or 
    not regiuni_list or 
    'sector_dominant' not in regiuni_list[0]):
    import csv
    with open('../datasets/regiuni_moldova.csv', 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        regiuni_list = []
        for row in reader:
            regiuni_list.append({
                'regiune': row['regiune'],
                'populatie': int(row['populatie']),
                'rata_ocupare': float(row['rata_ocupare']),
                'venit_mediu': float(row['venit_mediu']),
                'rata_alfabetizare': float(row['rata_alfabetizare']),
                'sector_dominant': row['sector_dominant'],
                'numar_intreprinderi': int(row['numar_intreprinderi']),
                'rata_somaj': float(row['rata_somaj'])
            })

# CURS 1 - destul de complex!
print("\nğŸ“ METODA CURS 1:")
sectoare = {}
for regiune in regiuni_list:
    sector = regiune['sector_dominant']
    if sector not in sectoare:
        sectoare[sector] = 0
    sectoare[sector] += regiune['populatie']

for sector, populatie in sectoare.items():
    print(f"  {sector}: {populatie:,}")
print("Linii de cod: ~10")

# PANDAS - incredibil de simplu!
print("\nâš¡ METODA PANDAS:")
print(df.groupby('sector_dominant')['populatie'].sum())
print("Linii de cod: 1")

print("\nğŸ“Š PANDAS (formatat mai frumos):")
rezultat = df.groupby('sector_dominant')['populatie'].sum().reset_index()
rezultat.columns = ['Sector', 'PopulaÈ›ie TotalÄƒ']
print(rezultat)
````

âœ… **Avantajele pandas:**
