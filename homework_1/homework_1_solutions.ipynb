### âœ… ExerciÈ›iul 1: CurÄƒÈ›area textului

```python
def curata_text(text):
    """
    CurÄƒÈ›Äƒ un È™ir de text eliminÃ¢nd spaÈ›iile È™i standardizÃ¢nd formatul.
    
    Parametri:
        text (str): Textul de curÄƒÈ›at
    
    ReturneazÄƒ:
        str: Textul curÄƒÈ›at
    
    Exemplu:
        curata_text("  agricultura  ") â†’ "Agricultura"
    """
    # Pasul 1: EliminÄƒm spaÈ›iile de la Ã®nceput È™i sfÃ¢rÈ™it cu .strip()
    text_fara_spatii = text.strip()
    
    # Pasul 2: Convertim la Title Case (prima literÄƒ din fiecare cuvÃ¢nt - majusculÄƒ)
    # .title() face: "agricultura" â†’ "Agricultura", "SERVICII" â†’ "Servicii"
    text_curat = text_fara_spatii.title()
    
    # ReturnÄƒm rezultatul final
    return text_curat

# Testare funcÈ›ie
print("=== Test curata_text ===")
print(f"'  Cahul' â†’ '{curata_text('  Cahul')}'")
print(f"'servicii' â†’ '{curata_text('servicii')}'")
print(f"'AGRICULTURÄ‚' â†’ '{curata_text('AGRICULTURÄ‚')}'")
```

### âœ… ExerciÈ›iul 2: Tratarea valorilor lipsÄƒ

```python
def calculeaza_medie_camp(lista_regiuni, nume_camp):
    """
    CalculeazÄƒ media unui cÃ¢mp numeric, ignorÃ¢nd valorile None.
    
    Parametri:
        lista_regiuni (list): Lista cu dicÈ›ionare de regiuni
        nume_camp (str): Numele cÃ¢mpului pentru care calculÄƒm media
    
    ReturneazÄƒ:
        float: Media valorilor (ignorÃ¢nd None)
    
    Exemplu:
        calculeaza_medie_camp(date_regiuni, "venit_mediu") â†’ 5845.0
    """
    # IniÈ›ializÄƒm variabilele pentru calcul
    suma = 0  # Suma totalÄƒ a valorilor
    count = 0  # CÃ¢te valori valide (non-None) am gÄƒsit
    
    # Parcurgem fiecare regiune din listÄƒ
    for regiune in lista_regiuni:
        # Extragem valoarea cÃ¢mpului specificat
        valoare = regiune[nume_camp]
        
        # VerificÄƒm dacÄƒ valoarea NU este None
        # 'is not None' este modul corect de a verifica Ã®n Python
        if valoare is not None:
            suma += valoare  # AdÄƒugÄƒm valoarea la sumÄƒ
            count += 1  # IncrementÄƒm contorul de valori valide
    
    # CalculÄƒm media: suma Ã®mpÄƒrÈ›itÄƒ la numÄƒrul de valori
    # IMPORTANT: count > 0 previne Ã®mpÄƒrÈ›irea la zero
    if count > 0:
        media = suma / count
    else:
        media = 0  # DacÄƒ nu avem nicio valoare validÄƒ, returnÄƒm 0
    
    return media

# Testare funcÈ›ie
print("\n=== Test calculeaza_medie_camp ===")
medie_venit = calculeaza_medie_camp(date_regiuni_raw, "venit_mediu")
print(f"Venit mediu (ignorÃ¢nd None): {medie_venit:.2f} lei")

medie_ocupare = calculeaza_medie_camp(date_regiuni_raw, "rata_ocupare")
print(f"Rata ocupare medie: {medie_ocupare:.2f}%")
```

### âœ… ExerciÈ›iul 3: Completarea valorilor lipsÄƒ

```python
def completeaza_valori_lipsa(lista_regiuni, nume_camp, valoare_inlocuire):
    """
    ÃnlocuieÈ™te toate valorile None dintr-un cÃ¢mp cu o valoare specificatÄƒ.
    
    Parametri:
        lista_regiuni (list): Lista cu dicÈ›ionare de regiuni (modificatÄƒ in-place)
        nume_camp (str): Numele cÃ¢mpului de verificat
        valoare_inlocuire: Valoarea cu care Ã®nlocuim None
    
    ReturneazÄƒ:
        list: Lista modificatÄƒ (aceeaÈ™i listÄƒ, nu o copie)
    """
    # Parcurgem fiecare regiune
    for regiune in lista_regiuni:
        # VerificÄƒm dacÄƒ cÃ¢mpul specificat este None
        if regiune[nume_camp] is None:
            # Ãnlocuim None cu valoarea de Ã®nlocuire
            regiune[nume_camp] = valoare_inlocuire
    
    # ReturnÄƒm lista (deÈ™i e modificatÄƒ in-place, e bine sÄƒ returnÄƒm pentru claritate)
    return lista_regiuni

# Vom testa Ã®n funcÈ›ia urmÄƒtoare care combinÄƒ toate operaÈ›iile
```

### âœ… ExerciÈ›iul 4: Proces complet de curÄƒÈ›are

```python
def curata_date_complete(lista_regiuni):
    """
    CurÄƒÈ›Äƒ complet un set de date: text, valori lipsÄƒ, format.
    
    PaÈ™i executaÈ›i:
    1. CreeazÄƒ o copie a datelor originale
    2. CurÄƒÈ›Äƒ toate cÃ¢mpurile text
    3. CalculeazÄƒ mediile pentru cÃ¢mpuri cu None
    4. CompleteazÄƒ valorile lipsÄƒ cu mediile
    
    Parametri:
        lista_regiuni (list): Lista originalÄƒ (rÄƒmÃ¢ne nemodificatÄƒ)
    
    ReturneazÄƒ:
        list: O listÄƒ nouÄƒ cu datele curÄƒÈ›ate
    """
    # Pasul 1: CreÄƒm o copie profundÄƒ (deep copy) a listei
    # Nu putem folosi doar .copy() pentru cÄƒ avem dicÈ›ionare Ã®n listÄƒ
    # Trebuie sÄƒ copiem È™i dicÈ›ionarele individual
    lista_curatata = []
    for regiune in lista_regiuni:
        # CreÄƒm un dicÈ›ionar nou cu aceleaÈ™i date
        regiune_noua = {}
        for key, value in regiune.items():
            regiune_noua[key] = value
        lista_curatata.append(regiune_noua)
    
    # Pasul 2: CurÄƒÈ›Äƒm toate cÃ¢mpurile text
    print("CurÄƒÈ›are cÃ¢mpuri text...")
    for regiune in lista_curatata:
        # CurÄƒÈ›Äƒm numele regiunii
        regiune["regiune"] = curata_text(regiune["regiune"])
        # CurÄƒÈ›Äƒm sectorul dominant
        regiune["sector_dominant"] = curata_text(regiune["sector_dominant"])
    
    # Pasul 3: IdentificÄƒm cÃ¢mpurile care pot avea None È™i calculÄƒm mediile
    print("Calculare medii pentru valori lipsÄƒ...")
    campuri_numerice = ["rata_ocupare", "venit_mediu", "rata_alfabetizare", "rata_somaj"]
    
    medii = {}  # DicÈ›ionar pentru a stoca mediile
    for camp in campuri_numerice:
        media = calculeaza_medie_camp(lista_curatata, camp)
        medii[camp] = media
        print(f"  {camp}: {media:.2f}")
    
    # Pasul 4: CompletÄƒm valorile lipsÄƒ cu mediile calculate
    print("\nCompletare valori lipsÄƒ...")
    valori_completate = 0
    for camp in campuri_numerice:
        # Pentru fiecare cÃ¢mp, Ã®nlocuim None cu media
        for regiune in lista_curatata:
            if regiune[camp] is None:
                regiune[camp] = medii[camp]
                valori_completate += 1
                print(f"  {regiune['regiune']}.{camp}: None â†’ {medii[camp]:.2f}")
    
    print(f"\nâœ“ CurÄƒÈ›are completÄƒ! {valori_completate} valori au fost completate.")
    
    return lista_curatata

# AplicÄƒm curÄƒÈ›area
print("=== PROCES DE CURÄ‚ÈšARE ===")
date_regiuni = curata_date_complete(date_regiuni_raw)

print("\n=== VERIFICARE DATE CURÄ‚ÈšATE ===")
print("\nPrimele 3 regiuni dupÄƒ curÄƒÈ›are:")
for i in range(3):
    print(f"\n{date_regiuni[i]['regiune']}:")
    print(f"  Sector: {date_regiuni[i]['sector_dominant']}")
    print(f"  Venit: {date_regiuni[i]['venit_mediu']:.2f} lei")
    print(f"  Rata ocupare: {date_regiuni[i]['rata_ocupare']:.2f}%")
```

### âœ… ExerciÈ›iul 7: Calcularea statisticilor naÈ›ionale

```python
def calculeaza_statistici_nationale(lista_regiuni):
    """
    CalculeazÄƒ statistici agregate la nivel naÈ›ional.
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
    
    ReturneazÄƒ:
        dict: DicÈ›ionar cu statistici naÈ›ionale
    """
    # IniÈ›ializÄƒm variabile pentru acumulare
    populatie_totala = 0
    suma_venituri = 0
    suma_rata_ocupare = 0
    suma_rata_somaj = 0
    suma_intreprinderi = 0
    
    # NumÄƒrul de regiuni (pentru calcul medii)
    numar_regiuni = len(lista_regiuni)
    
    # Parcurgem toate regiunile È™i acumulÄƒm valorile
    for regiune in lista_regiuni:
        populatie_totala += regiune["populatie"]
        suma_venituri += regiune["venit_mediu"]
        suma_rata_ocupare += regiune["rata_ocupare"]
        suma_rata_somaj += regiune["rata_somaj"]
        suma_intreprinderi += regiune["numar_intreprinderi"]
    
    # CalculÄƒm mediile
    venit_mediu_national = suma_venituri / numar_regiuni
    rata_ocupare_medie = suma_rata_ocupare / numar_regiuni
    rata_somaj_medie = suma_rata_somaj / numar_regiuni
    
    # CreÄƒm È™i returnÄƒm dicÈ›ionarul cu rezultate
    statistici = {
        "numar_regiuni": numar_regiuni,
        "populatie_totala": populatie_totala,
        "venit_mediu_national": venit_mediu_national,
        "rata_ocupare_medie": rata_ocupare_medie,
        "rata_somaj_medie": rata_somaj_medie,
        "numar_total_intreprinderi": suma_intreprinderi
    }
    
    return statistici

# Testare
print("\n=== Test calculeaza_statistici_nationale ===")
statistici = calculeaza_statistici_nationale(date_regiuni)
print("\nğŸ“Š STATISTICI NAÈšIONALE:")
print(f"  NumÄƒr regiuni:        {statistici['numar_regiuni']}")
print(f"  PopulaÈ›ie totalÄƒ:     {statistici['populatie_totala']:,} locuitori")
print(f"  Venit mediu naÈ›ional: {statistici['venit_mediu_national']:,.2f} lei")
print(f"  Rata ocupare medie:   {statistici['rata_ocupare_medie']:.2f}%")
print(f"  Rata È™omaj medie:     {statistici['rata_somaj_medie']:.2f}%")
print(f"  Total Ã®ntreprinderi:  {statistici['numar_total_intreprinderi']:,}")
```

### âœ… ExerciÈ›iul 8: AdÄƒugarea de cÃ¢mpuri calculate

```python
def adauga_densitate_intreprinderi(lista_regiuni):
    """
    AdaugÄƒ un cÃ¢mp calculat: densitatea Ã®ntreprinderilor per 1000 locuitori.
    
    Formula: densitate = numar_intreprinderi / (populatie / 1000)
    
    Parametri:
        lista_regiuni (list): Lista de regiuni (modificatÄƒ in-place)
    
    ReturneazÄƒ:
        list: Lista modificatÄƒ cu noul cÃ¢mp
    """
    # Parcurgem fiecare regiune
    for regiune in lista_regiuni:
        # Extragem datele necesare
        numar_intreprinderi = regiune["numar_intreprinderi"]
        populatie = regiune["populatie"]
        
        # CalculÄƒm densitatea
        # ÃmpÄƒrÈ›im populaÈ›ia la 1000 pentru a obÈ›ine numÄƒrul de "mii de locuitori"
        # Apoi Ã®mpÄƒrÈ›im numÄƒrul de Ã®ntreprinderi la acest rezultat
        densitate = numar_intreprinderi / (populatie / 1000)
        
        # AdÄƒugÄƒm noul cÃ¢mp Ã®n dicÈ›ionar
        regiune["densitate_intreprinderi"] = densitate
    
    return lista_regiuni

# AplicÄƒm funcÈ›ia
date_regiuni = adauga_densitate_intreprinderi(date_regiuni)

# Testare - afiÈ™Äƒm top 5 dupÄƒ densitate
print("\n=== Test adauga_densitate_intreprinderi ===")
print("\nDensitatea Ã®ntreprinderilor pe regiuni:")
for regiune in date_regiuni:
    print(f"  {regiune['regiune']:12} {regiune['densitate_intreprinderi']:5.2f} Ã®ntreprinderi/1000 loc.")
```

### âœ… ExerciÈ›iul 9: Filtrare cu multiple criterii

```python
def filtreaza_regiuni_dezvoltate(lista_regiuni):
    """
    FiltreazÄƒ regiunile dezvoltate pe baza mai multor criterii.
    
    Criterii (TOATE trebuie Ã®ndeplinite):
    - Venit mediu > 5500 lei
    - Rata ocupare > 60%
    - Rata alfabetizare > 98%
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
    
    ReturneazÄƒ:
        list: Liste cu numele regiunilor dezvoltate
    """
    regiuni_dezvoltate = []  # ListÄƒ pentru rezultat
    
    # Parcurgem fiecare regiune
    for regiune in lista_regiuni:
        # VerificÄƒm toate cele 3 criterii folosind 'and'
        # TOATE condiÈ›iile trebuie sÄƒ fie True
        if (regiune["venit_mediu"] > 5500 and
            regiune["rata_ocupare"] > 60 and
            regiune["rata_alfabetizare"] > 98):
            
            # DacÄƒ toate condiÈ›iile sunt Ã®ndeplinite, adÄƒugÄƒm regiunea
            regiuni_dezvoltate.append(regiune["regiune"])
    
    return regiuni_dezvoltate

# Testare
print("\n=== Test filtreaza_regiuni_dezvoltate ===")
regiuni_dev = filtreaza_regiuni_dezvoltate(date_regiuni)
print(f"\nRegiuni dezvoltate (toate criteriile Ã®ndeplinite): {len(regiuni_dev)}")
for regiune in regiuni_dev:
    print(f"  âœ“ {regiune}")
```

### âœ… ExerciÈ›iul 10: Clasificarea regiunilor

```python
def clasifica_regiune_dupa_populatie(populatie):
    """
    ClasificÄƒ o regiune Ã®n funcÈ›ie de populaÈ›ia sa.
    
    Parametri:
        populatie (int): PopulaÈ›ia regiunii
    
    ReturneazÄƒ:
        str: Categoria regiunii
    """
    # Folosim if-elif pentru verificÄƒri Ã®n cascadÄƒ
    # Ãncepem cu valorile cele mai mari
    if populatie > 100000:
        return "OraÈ™ mare"
    elif populatie >= 30000:  # Ã®ntre 30,000 È™i 100,000
        return "OraÈ™ mediu"
    elif populatie >= 15000:  # Ã®ntre 15,000 È™i 30,000
        return "OraÈ™ mic"
    else:  # sub 15,000
        return "Localitate micÄƒ"

def adauga_clasificare_dimensiune(lista_regiuni):
    """
    AdaugÄƒ cÃ¢mpul 'dimensiune' pentru fiecare regiune.
    
    Parametri:
        lista_regiuni (list): Lista de regiuni (modificatÄƒ in-place)
    
    ReturneazÄƒ:
        list: Lista modificatÄƒ
    """
    # Parcurgem fiecare regiune
    for regiune in lista_regiuni:
        # Folosim funcÈ›ia anterioarÄƒ pentru clasificare
        populatie = regiune["populatie"]
        dimensiune = clasifica_regiune_dupa_populatie(populatie)
        
        # AdÄƒugÄƒm noul cÃ¢mp
        regiune["dimensiune"] = dimensiune
    
    return lista_regiuni

# AplicÄƒm clasificarea
date_regiuni = adauga_clasificare_dimensiune(date_regiuni)

# Testare
print("\n=== Test clasificare dimensiune ===")
print("\nRegiuni clasificate dupÄƒ dimensiune:")
for regiune in date_regiuni:
    print(f"  {regiune['regiune']:12} - {regiune['dimensiune']:16} ({regiune['populatie']:>7,} loc.)")

# NumÄƒrÄƒm cÃ¢te regiuni sunt Ã®n fiecare categorie
print("\nDistribuÈ›ie pe categorii:")
categorii = {}
for regiune in date_regiuni:
    dim = regiune["dimensiune"]
    if dim not in categorii:
        categorii[dim] = 0
    categorii[dim] += 1

for categorie, numar in categorii.items():
    print(f"  {categorie}: {numar} regiuni")
```

### âœ… ExerciÈ›iul 11: Gruparea pe sectoare

```python
def grupeaza_dupa_sector(lista_regiuni):
    """
    GrupeazÄƒ regiunile dupÄƒ sectorul economic dominant.
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
    
    ReturneazÄƒ:
        dict: DicÈ›ionar {sector: [listÄƒ_regiuni]}
    """
    grupare = {}  # DicÈ›ionar pentru rezultat
    
    # Parcurgem fiecare regiune
    for regiune in lista_regiuni:
        sector = regiune["sector_dominant"]
        nume_regiune = regiune["regiune"]
        
        # VerificÄƒm dacÄƒ sectorul existÄƒ deja Ã®n dicÈ›ionar
        if sector not in grupare:
            # DacÄƒ nu existÄƒ, creÄƒm o listÄƒ nouÄƒ
            grupare[sector] = []
        
        # AdÄƒugÄƒm regiunea la lista sectorului corespunzÄƒtor
        grupare[sector].append(nume_regiune)
    
    return grupare

# Testare
print("\n=== Test grupeaza_dupa_sector ===")
grupare = grupeaza_dupa_sector(date_regiuni)

print("\nğŸ­ REGIUNI GRUPATE PE SECTOARE ECONOMICE:")
for sector, regiuni in grupare.items():
    print(f"\n{sector} ({len(regiuni)} regiuni):")
    for regiune in regiuni:
        print(f"  â€¢ {regiune}")
```

### âœ… ExerciÈ›iul 12: Statistici avansate pe sectoare

```python
def calculeaza_statistici_sector(lista_regiuni, sector):
    """
    CalculeazÄƒ statistici detaliate pentru un sector economic specific.
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
        sector (str): Numele sectorului de analizat
    
    ReturneazÄƒ:
        dict: DicÈ›ionar cu statistici despre sector
    """
    # Variabile pentru acumulare
    numar_regiuni = 0
    populatie_totala = 0
    suma_venituri = 0
    suma_rata_ocupare = 0
    suma_rata_somaj = 0
    
    # Parcurgem toate regiunile È™i selectÄƒm doar cele din sectorul dorit
    for regiune in lista_regiuni:
        if regiune["sector_dominant"] == sector:
            # AceastÄƒ regiune aparÈ›ine sectorului
            numar_regiuni += 1
            populatie_totala += regiune["populatie"]
            suma_venituri += regiune["venit_mediu"]
            suma_rata_ocupare += regiune["rata_ocupare"]
            suma_rata_somaj += regiune["rata_somaj"]
    
    # CalculÄƒm mediile (verificÄƒm Ã®mpÄƒrÈ›irea la zero)
    if numar_regiuni > 0:
        populatie_medie = populatie_totala / numar_regiuni
        venit_mediu = suma_venituri / numar_regiuni
        rata_ocupare_medie = suma_rata_ocupare / numar_regiuni
        rata_somaj_medie = suma_rata_somaj / numar_regiuni
    else:
        populatie_medie = 0
        venit_mediu = 0
        rata_ocupare_medie = 0
        rata_somaj_medie = 0
    
    # Construim dicÈ›ionarul cu rezultate
    statistici = {
        "sector": sector,
        "numar_regiuni": numar_regiuni,
        "populatie_totala": populatie_totala,
        "populatie_medie": populatie_medie,
        "venit_mediu": venit_mediu,
        "rata_ocupare_medie": rata_ocupare_medie,
        "rata_somaj_medie": rata_somaj_medie
    }
    
    return statistici

# Testare pentru toate sectoarele
print("\n=== Test calculeaza_statistici_sector ===")
grupare = grupeaza_dupa_sector(date_regiuni)

print("\nğŸ“ˆ STATISTICI DETALIATE PE SECTOARE:\n")
for sector in grupare.keys():
    stats = calculeaza_statistici_sector(date_regiuni, sector)
    
    print(f"{'='*50}")
    print(f"  {sector}")
    print(f"{'='*50}")
    print(f"  NumÄƒr regiuni:        {stats['numar_regiuni']}")
    print(f"  PopulaÈ›ie totalÄƒ:     {stats['populatie_totala']:,} locuitori")
    print(f"  PopulaÈ›ie medie:      {stats['populatie_medie']:,.0f} locuitori")
    print(f"  Venit mediu:          {stats['venit_mediu']:,.2f} lei")
    print(f"  Rata ocupare medie:   {stats['rata_ocupare_medie']:.2f}%")
    print(f"  Rata È™omaj medie:     {stats['rata_somaj_medie']:.2f}%")
    print()
```

### âœ… ExerciÈ›iul 13: GÄƒsirea extremelor

```python
def gaseste_extreme(lista_regiuni, camp, tip="max"):
    """
    GÄƒseÈ™te regiunea cu valoarea maximÄƒ sau minimÄƒ pentru un cÃ¢mp dat.
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
        camp (str): Numele cÃ¢mpului de analizat
        tip (str): "max" pentru maxim, "min" pentru minim
    
    ReturneazÄƒ:
        dict: {"regiune": nume, "valoare": valoare}
    """
    # VerificÄƒm cÄƒ avem date
    if len(lista_regiuni) == 0:
        return {"regiune": None, "valoare": None}
    
    # IniÈ›ializÄƒm cu prima regiune
    regiune_extrema = lista_regiuni[0]["regiune"]
    valoare_extrema = lista_regiuni[0][camp]
    
    # Parcurgem restul regiunilor (Ã®ncepÃ¢nd de la indexul 1)
    for regiune in lista_regiuni[1:]:
        valoare_curenta = regiune[camp]
        
        # ComparÄƒm Ã®n funcÈ›ie de tipul cerut
        if tip == "max":
            # CÄƒutÄƒm maximum - actualizÄƒm dacÄƒ gÄƒsim valoare mai mare
            if valoare_curenta > valoare_extrema:
                valoare_extrema = valoare_curenta
                regiune_extrema = regiune["regiune"]
        else:  # tip == "min"
            # CÄƒutÄƒm minimum - actualizÄƒm dacÄƒ gÄƒsim valoare mai micÄƒ
            if valoare_curenta < valoare_extrema:
                valoare_extrema = valoare_curenta
                regiune_extrema = regiune["regiune"]
    
    # ReturnÄƒm rezultatul
    return {
        "regiune": regiune_extrema,
        "valoare": valoare_extrema
    }

# Testare
print("\n=== Test gaseste_extreme ===")
print("\nğŸ” EXTREME PENTRU DIFERIÈšI INDICATORI:\n")

# Venit
max_venit = gaseste_extreme(date_regiuni, "venit_mediu", "max")
min_venit = gaseste_extreme(date_regiuni, "venit_mediu", "min")
print(f"Venit:")
print(f"  â†‘ Maxim: {max_venit['regiune']} ({max_venit['valoare']:,.0f} lei)")
print(f"  â†“ Minim: {min_venit['regiune']} ({min_venit['valoare']:,.0f} lei)")

# PopulaÈ›ie
max_pop = gaseste_extreme(date_regiuni, "populatie", "max")
min_pop = gaseste_extreme(date_regiuni, "populatie", "min")
print(f"\nPopulaÈ›ie:")
print(f"  â†‘ Maxim: {max_pop['regiune']} ({max_pop['valoare']:,} locuitori)")
print(f"  â†“ Minim: {min_pop['regiune']} ({min_pop['valoare']:,} locuitori)")

# È˜omaj
max_somaj = gaseste_extreme(date_regiuni, "rata_somaj", "max")
min_somaj = gaseste_extreme(date_regiuni, "rata_somaj", "min")
print(f"\nÈ˜omaj:")
print(f"  â†‘ Maxim: {max_somaj['regiune']} ({max_somaj['valoare']:.1f}%)")
print(f"  â†“ Minim: {min_somaj['regiune']} ({min_somaj['valoare']:.1f}%)")
```

```python
def top_regiuni(lista_regiuni, camp, n=5, descrescator=True):
    """
    ReturneazÄƒ top N regiuni sortate dupÄƒ un cÃ¢mp.
    
    AceastÄƒ implementare foloseÈ™te o sortare simplÄƒ (bubble sort adaptat).
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
        camp (str): CÃ¢mpul dupÄƒ care sortÄƒm
        n (int): CÃ¢te regiuni sÄƒ returnÄƒm
        descrescator (bool): True pentru descrescÄƒtor, False pentru crescÄƒtor
    
    ReturneazÄƒ:
        list: ListÄƒ de dicÈ›ionare {"regiune": nume, "valoare": valoare}
    """
    # Pasul 1: CreÄƒm o listÄƒ de perechi (nume_regiune, valoare)
    perechi = []
    for regiune in lista_regiuni:
        perechi.append({
            "regiune": regiune["regiune"],
            "valoare": regiune[camp]
        })
    
    # Pasul 2: SortÄƒm lista folosind bubble sort simplificat
    # Parcurgem lista de mai multe ori
    for i in range(len(perechi)):
        for j in range(i + 1, len(perechi)):
            # ComparÄƒm elementul i cu elementul j
            if descrescator:
                # Pentru descrescÄƒtor: dacÄƒ i < j, le inversÄƒm
                if perechi[i]["valoare"] < perechi[j]["valoare"]:
                    # Swap (interschimbÄƒm)
                    temp = perechi[i]
                    perechi[i] = perechi[j]
                    perechi[j] = temp
            else:
                # Pentru crescÄƒtor: dacÄƒ i > j, le inversÄƒm
                if perechi[i]["valoare"] > perechi[j]["valoare"]:
                    temp = perechi[i]
                    perechi[i] = perechi[j]
                    perechi[j] = temp
    
    # Pasul 3: ReturnÄƒm primele N elemente
    # Folosim slicing pentru a lua primele n elemente
    top_n = perechi[:n]
    
    return top_n

# Testare
print("\n=== Test top_regiuni ===")

print("\nğŸ† TOP 5 REGIUNI DUPÄ‚ POPULAÈšIE:")
top_pop = top_regiuni(date_regiuni, "populatie", 5)
for i, regiune in enumerate(top_pop, 1):
    print(f"  {i}. {regiune['regiune']:12} - {regiune['valoare']:>7,} locuitori")

print("\nğŸ’° TOP 5 REGIUNI DUPÄ‚ VENIT MEDIU:")
top_venit = top_regiuni(date_regiuni, "venit_mediu", 5)
for i, regiune in enumerate(top_venit, 1):
    print(f"  {i}. {regiune['regiune']:12} - {regiune['valoare']:>7,.0f} lei")

print("\nğŸ¢ TOP 5 REGIUNI DUPÄ‚ DENSITATE ÃNTREPRINDERI:")
top_densitate = top_regiuni(date_regiuni, "densitate_intreprinderi", 5)
for i, regiune in enumerate(top_densitate, 1):
    print(f"  {i}. {regiune['regiune']:12} - {regiune['valoare']:>5.2f}/1000 loc.")
```

```python
def compara_regiuni(lista_regiuni, nume1, nume2):
    """
    AfiÈ™eazÄƒ o comparaÈ›ie detaliatÄƒ Ã®ntre douÄƒ regiuni.
    
    Parametri:
        lista_regiuni (list): Lista cu datele regiunilor
        nume1, nume2 (str): Numele celor douÄƒ regiuni de comparat
    """
    # Pasul 1: GÄƒsim cele douÄƒ regiuni Ã®n listÄƒ
    regiune1 = None
    regiune2 = None
    
    for regiune in lista_regiuni:
        if regiune["regiune"] == nume1:
            regiune1 = regiune
        if regiune["regiune"] == nume2:
            regiune2 = regiune
    
    # VerificÄƒm dacÄƒ am gÄƒsit ambele regiuni
    if regiune1 is None:
        print(f"Eroare: Regiunea '{nume1}' nu a fost gÄƒsitÄƒ!")
        return
    if regiune2 is None:
        print(f"Eroare: Regiunea '{nume2}' nu a fost gÄƒsitÄƒ!")
        return
    
    # Pasul 2: AfiÈ™Äƒm comparaÈ›ia
    print(f"\n{'='*60}")
    print(f"  COMPARAÈšIE: {nume1} vs {nume2}")
    print(f"{'='*60}\n")
    
    # FuncÈ›ie helper pentru comparare È™i afiÈ™are
    def compara_indicator(nume_indicator, valoare1, valoare2, unitate=""):
        print(f"{nume_indicator}:")
        print(f"  {nume1:12} {valoare1:>12,.2f} {unitate}")
        print(f"  {nume2:12} {valoare2:>12,.2f} {unitate}")
        
        # CalculÄƒm diferenÈ›a
        diferenta = valoare1 - valoare2
        if diferenta > 0:
            print(f"  â†’ {nume1} are cu {abs(diferenta):,.2f} {unitate} mai mult")
        elif diferenta < 0:
            print(f"  â†’ {nume2} are cu {abs(diferenta):,.2f} {unitate} mai mult")
        else:
            print(f"  â†’ Valori egale")
        print()
    
    # ComparÄƒm toÈ›i indicatorii
    compara_indicator("PopulaÈ›ie", regiune1["populatie"], regiune2["populatie"], "locuitori")
    compara_indicator("Venit mediu", regiune1["venit_mediu"], regiune2["venit_mediu"], "lei")
    compara_indicator("Rata ocupare", regiune1["rata_ocupare"], regiune2["rata_ocupare"], "%")
    compara_indicator("Rata È™omaj", regiune1["rata_somaj"], regiune2["rata_somaj"], "%")
    compara_indicator("Rata alfabetizare", regiune1["rata_alfabetizare"], regiune2["rata_alfabetizare"], "%")
    compara_indicator("Ãntreprinderi", regiune1["numar_intreprinderi"], regiune2["numar_intreprinderi"], "")
    
    # AfiÈ™Äƒm È™i diferenÈ›ele calitative
    print("DiferenÈ›e calitative:")
    print(f"  Sector dominant: {regiune1['sector_dominant']} vs {regiune2['sector_dominant']}")
    print(f"  Dimensiune: {regiune1['dimensiune']} vs {regiune2['dimensiune']}")

# Testare
print("\n=== Test compara_regiuni ===")
compara_regiuni(date_regiuni, "ChiÈ™inÄƒu", "BÄƒlÈ›i")
compara_regiuni(date_regiuni, "EdineÈ›", "Comrat")
```
